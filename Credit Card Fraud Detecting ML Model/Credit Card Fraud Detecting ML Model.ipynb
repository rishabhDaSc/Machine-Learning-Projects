```python
# IMPORTING NECESSARY LIBRARIES
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# IMPORTING DATASET
data = pd.read_csv('creditcard.csv')
```

```python
data.head()
```

```python
# CHECKING NULL VALUES
data.isnull().sum().sum()
```

```python
# SHAPE OF DATA
data.shape
```

```python
# CHECKING DUPLICATE DATA
data[data.duplicated()]
```

```python
# DROPING DUPLICATE DATA
data.drop_duplicates(inplace=True)
```

```python
# SHAPE OF DATA AFTER DROPING DUPLICATE VALUES
data.shape
```

```python
# STATISTICAL ANALYSIS
data.describe()
```

```python
# NUMBER OF FRAUD AND LEGIT TRANSACTIONS
data.Class.value_counts()
```

```python
# DATA PREPROCESSING
data['Class'] = data['Class'].replace(to_replace=[0,1],value=['Legit','Fraud'])
```

```python
data.Class.value_counts()
```

```python
Legit = data[data['Class']=='Legit']
Fraud = data[data['Class']=='Fraud']
```

```python
print(Legit.shape)
print(Fraud.shape)
```

```python
# REDUCING UNBALANCED DATA TO EQUAL 
Legit_1 = Legit.sample(473,random_state=4)
```

```python
new_data = pd.concat([Legit_1,Fraud],axis=0,ignore_index=True)
```

```python
new_data.Class.value_counts()
```

```python
data.groupby('Class').mean()
```

```python
new_data.groupby('Class').mean()
```

```python
# SEPARATING FEATURES AND TARGET VARIABLES
X = new_data.drop(columns='Class')
Y = new_data['Class']
```

```python
# SPLITTING DATA TO TRAIN AND TEST
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=.25,random_state=1)
```

```python
# ML MODEL FITTING MODEL WITH TRAIN DATA
LR = LogisticRegression()
LR_model = LR.fit(x_train,y_train)
```

```python
# MODEL PREDICTION
y_train_pred = LR_model.predict(x_train)
y_test_pred = LR_model.predict(x_test)
```

```python
# MODEL ACCURACY
print(f'Model accuracy on Training data: {round(accuracy_score(y_train,y_train_pred)*100,2)}%')
print(f'Model accuracy on Test data: {round(accuracy_score(y_test,y_test_pred)*100,2)}%')
```

```python
# MODEL EVALUATION
Fraud_list = list(Y[Y=='Fraud'].index)
Legit_list = list(Y[Y=='Legit'].index)
Predicted_Fraud = []
Predicted_Legit = []
```

```python
for i in Fraud_list:
    Fr = X.iloc[i].values.reshape(1,-1)
    Fr_ = LR_model.predict(Fr)
    Predicted_Fraud.append(Fr_)
for j in Legit_list:
    Le = X.iloc[j].values.reshape(1,-1)
    Le_ = LR_model.predict(Le)
    Predicted_Legit.append(Le_)
```

```python
Fraud = pd.Series(Predicted_Fraud)
Legit = pd.Series(Predicted_Legit)
```

```python
print(f'Out of {len(Fraud_list)} FRAUD TRANSACTION DATA Logistic Regression Model Predicted')
print(Fraud.value_counts().to_string())
```

```python
print(f'Out of {len(Legit_list)} LEGIT TRANSACTION DATA Logistic Regression Model Predicted')
print(Legit.value_counts().to_string())
```
