```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.DataFrame(load_boston().data,columns=load_boston().feature_names)
data.head()
```

```python
data['Target'] = load_boston().target
data.head()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data.corr(),annot=True,fmt='.2f',)
```

```python
X = data.drop(columns=['Target'],axis=1)
Y = data['Target']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,random_state=3,test_size=.2)
```

```python
type(y_train)
```

```python
model = XGBRegressor()
model.fit(X_train,y_train)
```

```python
# Prediction on Training Data
X_train_pred = model.predict(X_train)
```

```python
print('R2 Score on Training Data',r2_score(y_train,X_train_pred))
print('Mean Absolute Error on Training Data',mean_absolute_error(y_train,X_train_pred))
print('Mean Squared Error on Training Data',mean_squared_error(y_train,X_train_pred))
```

```python
# Values of predicted values on training data and actual values
print(X_train_pred[:5])
print(list(y_train[:5]))
```

```python
# Prediction on Test data
X_test_pred = model.predict(X_test)
```

```python
print('R2 Score on Test Data',r2_score(y_test,X_test_pred))
print('Mean Absolute Error on Test Data',mean_absolute_error(y_test,X_test_pred))
print('Mean Squared Error on Test Data',mean_squared_error(y_test,X_test_pred))
```

```python
# Values of predicted values on test data and actual values
print(X_test_pred[:5])
print(list(y_test[:5]))
```
